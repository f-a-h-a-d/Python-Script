```python
import requests
import pandas as pd
import json
import mysql.connector as mysql
from sqlalchemy import create_engine
```

```python
response = requests.get("https://api.first.org/data/v1/countries")

text = response.json()['data']
text = json.dumps(text);

#json text to python dict

data = json.loads(text)

total_list = []

for key in data:
    list = []
    list.append('Short_notation')
    list.append(key)
    for k in data[key]:
       
        list.append(k)
        list.append(data[key][k])
    print(list)
    total_list.append(list)
```

```python
def creating_table(table_name):

        cursor = connection.cursor()
    
        query = f"CREATE TABLE IF NOT EXISTS {table_name} (Short_notation varchar(255), country varchar(255), region varchar(255))"
    
        print(query)
    
        cursor.execute(query)
        connection.commit()
        
            
        print(f"Successfuly create table {table_name}")   
    
creating_table("All_country")



```

```python
def inserting_data(list_of_list):

    cursor = connection.cursor()
    
    for val in list_of_list:
         
        cnt = 0
        query_text = "("
        fcnt = 0
        
        for fval in val:

            escape_val = connection._cmysql.escape_string(str(fval))
            escape_val = escape_val.decode('utf-8')
            
              
            if(cnt % 2 ==0):
                cnt = cnt + 1
                continue
            
            if(fcnt == 2):
                query_text += f"'{escape_val}'"
            else:
                query_text += f"'{escape_val}',"
                
            fcnt = fcnt + 1
            cnt = cnt + 1
            
        query_text += ")"

        main_query = f"INSERT INTO All_country VALUES {query_text}"
        print(main_query)

        cursor.execute(main_query)
        connection.commit()

        


inserting_data(total_list)
```
