```python
import random
import csv
import datetime
import string
import pandas as pd
```

```python
df = pd.read_csv('train.csv')

print(df)
```

```python
def date_order_reverse(str):
    str += '/'
    new_str = ""
    total = []
    for i in str:
        if i == '/':
            total.append(new_str)
            new_str = ""
            continue
            
        new_str += i
    
    total.reverse()
    returning = ""
    for val in total:
        returning += val
        returning +='/'

    returning= returning.rstrip(returning[-1])
    return returning

date_order_reverse('21/05/2017')
```

```python



def dateFix(value_of_date):
    value_of_date = date_order_reverse(value_of_date)
    new_value_of_date = value_of_date.replace('/', '-')
    return new_value_of_date

def orderIdFix(value_of_order):
    new_value_of_order = value_of_order.replace('-', '_')
    return new_value_of_order

def preProcessing():
    for i in df.index:
        fixedDateOrder = dateFix(df['Order Date'][i])
        print(fixedDateOrder)
        df.loc[i, 'Order Date'] = fixedDateOrder + " 00:00:00"

        fixedDateShip = dateFix(df['Ship Date'][i])
        df.loc[i, 'Ship Date'] = fixedDateShip + " 00:00:00"

        fixedOrderID = orderIdFix(df['Order ID'][i])
        df.loc[i, 'Order ID'] = fixedOrderID

        cusID = orderIdFix(df['Customer ID'][i])
        df.loc[i, 'Customer ID'] = cusID

    for j in range(18):
        df.columns.values[j] = orderIdFix(df.columns[j])
        
        

preProcessing()

df.to_csv('train_output.csv', mode = 'w' , index = False)


```
