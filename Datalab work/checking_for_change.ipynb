```python
response = requests.get("https://api.first.org/data/v1/countries")

text = response.json()['data']
text = json.dumps(text);

#json text to python dict

data = json.loads(text)

total_country = []

for key in data:
    for val in data[key]:
        value = data[key][val]
        value = value.replace("(the)", "")
        value = value.strip(" ")
        value = value.replace(" ", '+')
        total_country.append(value)
        
total_country.append("United+States")
print(total_country)
```

```python
def inserting_data(list_of_list):

    cursor = connection.cursor()
    
    for val in tqdm(list_of_list):
         
        query_text = "("
        
        for index, fval in enumerate(val):
            
            if(type(fval) == type([])):
                fval = json.dumps(fval)

            fval = connection._cmysql.escape_string(str(fval))
            fval = fval.decode('utf-8')

            if index == 5:
                query_text += f"'{fval}',"
            else:
                query_text += f"'{fval}',"

        query_text += f"'{datetime.now()}', '{datetime.now()}'"
            
        query_text += ")"

        main_query = f"INSERT INTO All_university VALUES {query_text}"
        print(main_query)

        cursor.execute(main_query)
        connection.commit()
```
